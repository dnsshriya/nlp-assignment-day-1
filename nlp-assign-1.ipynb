```python
df.drop("file", axis=1, inplace=True)
```

```python
import string

df["text_lower"]=df["message"].str.lower()
df.head()
```

```python

PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(data):
    """custom function to remove the punctuation"""
    return data.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

df["text_wo_punct"] = df["text_lower"].apply(lambda data: remove_punctuation(data))
df.head()
```

```python
from nltk.corpus import stopwords
", ".join(stopwords.words('english'))
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(data):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(data).split() if word not in STOPWORDS])

df["text_wo_stop"] = df["text_wo_punct"].apply(lambda data: remove_stopwords(data))
df.head()
```

```python
def check_word_in_stopwords(word):
    if word in set(stopwords.words('english')):
        return "Available"
```

```python
check_word_in_stopwords("was")
```

```python
from nltk.stem.porter import PorterStemmer
 

stemmer = PorterStemmer()
def stem_words(data):
    return " ".join([stemmer.stem(word) for word in data.split()])

df["text_stemmed"] = df["text_wo_stop"].apply(lambda data: stem_words(data))
df.head()
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
def lemmatize_words(data):
    return " ".join([lemmatizer.lemmatize(word) for word in data.split()])

df["text_lemmatized"] = df["text"].apply(lambda text: lemmatize_words(data))
df.head()
```
